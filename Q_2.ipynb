```python
import pandas as pd
```

```python
i_p = [0.114,0.113,0.112,.111,.099,.089,.079,.069,.059,.049,.039,.029,.019,.009,.006,.004]
```

```python
def find(probs,rem_probs,rest_p):
    for i in range(11):
        #print(probs)
        nonzeros = [j for j,e in enumerate(probs) if e!=0]
        team = min(nonzeros)
        #print(team)
        rem_probs[i][team] = rem_probs[i][team]+rest_p
        #print(rem_probs)
        probs[team] = 0
        
    return rem_probs
```

```python
import numpy
from numpy import *
s_p = [0.]*16
t_p = [0.]*16
f_p = [0.]*16
fi_p = [0.]*16
#r_probs = [[0 for i in range(16)] for j in range(11)]
r_probs = [[0]*16 for _ in range(11)]
for i in range(16):
    first_removed = i_p[:i] + [0.] + i_p[i+1:]
    #print(first_removed)
    total_prob = sum(first_removed)
    c_p = [i_p[i]*x/total_prob for x in first_removed]
    #print(c_p[0])
    s_p = [float("{0:.5f}".format(x+y)) for x,y in zip(s_p,c_p)]
    
    for j in range(16):
        if i!=j:
            second = first_removed[:j]+[0.]+first_removed[j+1:]
            total = sum(second)
            cp_2 = [c_p[j]*x/total for x in second]
            t_p = [float("{0:.5f}".format(x+y)) for x,y in zip(t_p,cp_2)]
            
            for k in range(16):
                if i!=k and j!=k:
                    third = second[:k]+[0.]+second[k+1:]
                    total_1 = sum(third)
                    cp_3 = [cp_2[k]*x/total_1 for x in third]
                    f_p = [float("{0:.5f}".format(x+y)) for x,y in zip(f_p,cp_3)]
                    
                    for l in range(16):
                        if i!= l and j!=l and k!=l:
                            fourth = third[:l]+[0.]+third[l+1:]
                            total_2 = sum(fourth)
                            cp_4 = [cp_3[l]*x/total_2 for x in fourth]
                            fi_p = [x+y for x,y in zip(fi_p,cp_4)]
                    
                            for m in range(16):
                                if i!=m and j!=m and k!=m and l!=m:
                                #if len(unique([i,j,k,l,m])) == 5:
                                        #print(cp_4[l])
                                        restProb = cp_4[m]
                                        #print(restProb)
                                        f_removed = fourth[:m]+[0.]+fourth[m+1:]
                                        #print(f_removed)
                                        r_probs = find(f_removed,r_probs,restProb)
```
